#Problem Statement 1:Smart Parking Lot Management System


```python
# Function to manage parking system

def manage_parking(capacity, vehicle_logs):

    current_vehicles = 0
    peak_usage = 0

    # Loop through vehicle logs
    for log in vehicle_logs:

        if log == "IN":
            current_vehicles += 1
        elif log == "OUT" and current_vehicles > 0:
            current_vehicles -= 1

        # Track peak usage
        if current_vehicles > peak_usage:
            peak_usage = current_vehicles

    print("Currently Parked Vehicles:", current_vehicles)

    # Check capacity condition
    if current_vehicles > capacity:
        print("Parking Status: Full - Capacity Exceeded")
    else:
        print("Parking Status: Available")


# Sample Input
manage_parking(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

#Problem Statement 2:Online Food Delivery Time Estimator


```python
# Function to estimate delivery time

def estimate_delivery_time(distance_km, traffic_level, weather_condition):

    # Base time calculation (5 minutes per km)
    total_time = distance_km * 5

    # Traffic adjustment
    if traffic_level.lower() == "high":
        total_time += 10
    elif traffic_level.lower() == "medium":
        total_time += 5

    # Weather adjustment
    if weather_condition.lower() == "rainy":
        total_time += 5
    elif weather_condition.lower() == "storm":
        total_time += 15

    return total_time


# Sample Input
eta = estimate_delivery_time(8, "High", "Rainy")
print("Estimated Delivery Time:", eta, "minutes")
```

# Problem Statement 3:Movie Theatre Seat Occupancy Analyzer

```python
# Movie Theatre Seat Occupancy Analyzer

def check_show_status(total_capacity, booking_list):

    # Count total booked seats using sum()
    # (Since booking_list contains 1 for each booked seat)
    booked_total = sum(booking_list)

    # Calculate occupancy percentage
    occupancy_rate = (booked_total / total_capacity) * 100

    # Decide show status based on occupancy
    if occupancy_rate >= 100:
        status_message = "Housefull"
    elif occupancy_rate >= 75:
        status_message = "Almost Full"
    else:
        status_message = "Seats Available"

    # Return rounded percentage and status
    return round(occupancy_rate), status_message


# Sample Input: 200 total seats, 150 booked
percentage, show_status = check_show_status(200, [1]*150)

# Display final result
print("Occupancy:", percentage, "%")
print("Show Status:", show_status)
```

# Problem Statement 4:Cloud Server Load Classification System


```python
# Cloud Server Load Classification System

def classify_server_load(cpu_readings):

    # Calculate total CPU usage using loop
    total_usage = 0
    for reading in cpu_readings:
        total_usage += reading

    # Calculate average CPU load
    average_load = total_usage / len(cpu_readings)

    # Print average CPU load
    print("Average CPU Load:", round(average_load), "%")

    # Classify server status based on average
    if average_load < 50:
        print("Server Status: Normal")
    elif average_load <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


# Sample Input
classify_server_load([45, 60, 70, 85, 90])
```

#Problem Statement 5:Smart Classroom Resource Usage Monitor


```python
# Classroom Energy Monitoring System

def evaluate_usage(usage_data):

    safe_limit = 8   # Maximum allowed usage hours

    # Identify resources that crossed the limit
    flagged = [name for name in usage_data
               if usage_data[name] > safe_limit]

    alert_status = "Yes" if flagged else "No"

    return flagged, alert_status


# Sample Input
resource_data = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

overuse_list, alert = evaluate_usage(resource_data)

print("Resources Exceeding Limit:",
      ", ".join(overuse_list) if overuse_list else "None")

print("Energy Alert:", alert)
```

# Problem Statement 6: Online Event Registration Capacity Controller


```python
# Online Event Registration Controller

def manage_event_capacity(max_capacity, total_requests):

    # Determine confirmed registrations
    confirmed_users = min(max_capacity, total_requests)

    # Calculate waitlisted users
    waitlist_users = 0
    if total_requests > max_capacity:
        waitlist_users = total_requests - max_capacity

    # Decide registration status
    registration_status = "Closed" if total_requests >= max_capacity else "Open"

    return confirmed_users, waitlist_users, registration_status


# Sample Input
capacity = 100
requests = 105

confirmed, waitlist, status = manage_event_capacity(capacity, requests)

print("Confirmed Registrations:", confirmed)
print("Waitlisted Users:", waitlist)
print("Registration Status:", status)
```
