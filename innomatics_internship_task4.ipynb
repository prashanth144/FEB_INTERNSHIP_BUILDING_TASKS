```python
# Problem 1: Mobile Recharge Validation System
#  Problem Statement
# Create a function to validate a mobile recharge request.
# Rules:
# Recharge amount must be ≥ ₹50


# Valid plans: 199, 299, 399, 599


# User should be allowed to retry if invalid
```

```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        print("Amount must be ₹50 or more")
        return False
    elif amount not in valid_plans:
        print("Invalid plan. Choose from:", valid_plans)
        return False
    else:
        print("Recharge of ₹", amount, "is successful!")
        return True


# Main Program
while True:
    amount = int(input("Enter recharge amount: "))
    result = validate_recharge(amount)

    if result == True:
        break
    else:
        retry = input("Try again? (yes/no): ")
        if retry != "yes":
            print("Cancelled.")
            break
```

```python
# Problem 2: Inventory Reorder Alert System
# Problem Statement
# Given a dictionary of products and stock quantities:
# Print Reorder Alert if stock < 15


# Print Stock OK otherwise
```

```python
def check_inventory(products):
    for product, stock in products.items():
        if stock < 15:
            print(product, "->", stock, "-> Reorder Alert!")
        else:
            print(product, "->", stock, "-> Stock OK")


# Sample Data
products = {
    "Laptop": 8,
    "Mouse": 25,
    "Keyboard": 12,
    "Monitor": 30,
    "USB Cable": 5
}

check_inventory(products)
```

```python
#  Problem 3: Student Result Processing System
# Problem Statement
# Create a function that:
# Accepts a list of student marks


# Calculates average marks


# Prints Pass if average ≥ 50 else Fail
```

```python
def process_result(marks):
    total = 0
    for mark in marks:
        total = total + mark

    average = total / len(marks)

    if average >= 50:
        return average, "Pass"
    else:
        return average, "Fail"


# Sample Data
marks = [78, 65, 42, 90, 55]

average, result = process_result(marks)
print("Average Marks:", average)
print("Result:", result)
```

```python
# Problem 4: Cab Fare Estimator with Retry Option
# Problem Statement
# Create a cab fare calculator:
# Base fare = ₹50


# ₹12 per km


# Peak hour → 25% extra


# Allow retry using while loop
```

```python
def calculate_fare(distance, peak):
    base_fare = 50
    rate = 12

    fare = base_fare + (rate * distance)

    if peak == "yes":
        extra = fare * 0.25
        fare = fare + extra

    return fare


# Main Program
while True:
    distance = float(input("Enter distance in km: "))
    peak = input("Peak hour? (yes/no): ")

    fare = calculate_fare(distance, peak)
    print("Total Fare: ₹", fare)

    retry = input("Calculate again? (yes/no): ")
    if retry != "yes":
        break
```

```python
# Problem 5: Employee Attendance Eligibility Checker
# Problem Statement
# Create a function that:
# Accepts attendance list ("P", "A")


# Calculates attendance percentage


# Returns Eligible if ≥ 75% else Not Eligible
```

```python
def check_attendance(attendance):
    total_days = len(attendance)
    present = 0

    for day in attendance:
        if day == "P":
            present = present + 1

    percentage = (present / total_days) * 100

    if percentage >= 75:
        return percentage, "Eligible"
    else:
        return percentage, "Not Eligible"


# Sample Data
attendance = ["P", "P", "A", "P", "P", "P", "A", "P", "P", "P"]

percentage, status = check_attendance(attendance)
print("Attendance:", percentage, "%")
print("Status:", status)
```

```python
# Problem 6: Password Strength Checker
# Problem Statement
# Create a function to check password strength.
# Rules:
# Minimum length = 8


# Must contain at least one digit


# Must contain at least one special character (@#$)
```

```python
def check_password(password):
    special = "@#$"
    has_digit = False
    has_special = False

    # Check length
    if len(password) < 8:
        print("Password too short. Minimum 8 characters needed.")
        return "Weak"

    # Check for digit and special character
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special:
            has_special = True

    if has_digit == False:
        print("Password must have at least one digit.")
        return "Weak"

    if has_special == False:
        print("Password must have at least one special character (@#$).")
        return "Weak"

    return "Strong"


# Test Passwords
passwords = ["hello", "hello123", "hello@world", "Hello@123"]

for pwd in passwords:
    print("\nPassword:", pwd)
    result = check_password(pwd)
    print("Strength:", result)
```
