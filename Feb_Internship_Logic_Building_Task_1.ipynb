TASK-2  Pass / Fail Analyzer


```python
marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("Pass:", pass_count)
print("Fail:", fail_count)

```

TASK-3  Simple Data Cleaner

```python
names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

for name in names:
    cleaned_name = name.strip().lower()   # Remove spaces and convert to lowercase
    cleaned_names.append(cleaned_name)

print("Cleaned Names:", cleaned_names)
```

TASK-4 Message Length Analyzer

```python
messages = ["Hi", "Welcome to the platform", "OK"]

for message in messages:
    length = len(message)
    print("Message:", message)
    print("Length:", length)

    if length > 10:
        print("This message is longer than 10 characters")


```

TASK-5  Error Message Detector

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
entries=[]
for log in logs:
  if log=="ERROR":
    entries.append(log)
print(len(entries))
```
