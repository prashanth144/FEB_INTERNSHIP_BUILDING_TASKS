Problem 1: Smart Light Controller


```python
class SmartLight:
    def __init__(self, light_name):
        self.light_name = light_name
        self.status = "OFF"  

    # Method to turn ON the light
    def turn_on(self):
        self.status = "ON"

    # Method to turn OFF the light
    def turn_off(self):
        self.status = "OFF"

    # Method to display current status
    def display_status(self):
        print(f"{self.light_name} is {self.status}")


light_name = input("Enter Light Name: ")
action = input("Enter Action (ON/OFF): ").strip().upper()

# Creating object of SmartLight class
light = SmartLight(light_name)

# Performing action based on input
if action == "ON":
    light.turn_on()
elif action == "OFF":
    light.turn_off()
else:
    print("Invalid Action! Please enter ON or OFF.")

light.display_status()
```

Problem 2: Employee ID Card System


```python
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    # Method to display ID Card details
    def display_id_card(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")


# ----------- Taking User Input -------------

name = input("Enter Employee Name: ")
emp_id = input("Enter Employee ID: ")
department = input("Enter Department: ")

employee = Employee(name, emp_id, department)

print()   
employee.display_id_card()
```

Problem 3: Mobile Contact Record


```python
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    # Method to display contact information
    def display_contact(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")


# ----------- Taking User Input -------------

name = input("Enter Contact Name: ")
phone = input("Enter Phone Number: ")

contact = Contact(name, phone)

print()  
contact.display_contact()
```

Problem 4: Product Price Tag Generator


```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    # Method to print formatted price tag
    def print_price_tag(self):
        print(f"Product: {self.name}")
        print(f"Price: ₹{self.price}")


# ----------- Taking User Input -------------

name = input("Enter Product Name: ")
price = input("Enter Price: ")

product = Product(name, price)

print()  
product.print_price_tag()
```

Problem 5: Movie Rating Display System


```python
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    # Method to display movie details
    def display_movie(self):
        print(f"Movie: {self.name}")
        print(f"Rating: {self.rating} / 5")


# ----------- Taking User Input -------------

name = input("Enter Movie Name: ")
rating = input("Enter Rating: ")

movie = Movie(name, rating)

print()  
movie.display_movie()
```

Problem 6: Delivery Address Manager


```python
class Delivery:
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    # Method to display delivery details
    def display_delivery(self):
        print("Delivery Details")
        print(f"Customer: {self.customer_name}")
        print(f"Address: {self.address}")


# ----------- Taking User Input -------------

customer_name = input("Enter Customer Name: ")
address = input("Enter Address: ")

delivery = Delivery(customer_name, address)

print()  
delivery.display_delivery()
```
